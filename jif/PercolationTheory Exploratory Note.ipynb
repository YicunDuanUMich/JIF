This notebook is meant to serve as the exploratory coding for a module that will eventually define local clusters of galaxies that should be simultaniously fit and extract a postage stamp around this region.

Some of the percolation theory code is based on:
http://dragly.org/2013/03/25/working-with-percolation-clusters-in-python/

Some of the image processing is based on:

http://scipy-lectures.github.io/advanced/image_processing/

Limitations/Assumptions of Current Process:
- Using a single circular Gaussian covariance to represent the ground PSF for the entire field. Ultimately we should use the infered PSF at the location of each galaxy and allow it to be non-circular and non-Gaussian.
- When convolving the space objects with the ground PSF, there is no correction for the space PSF.
- There is no correction for differing depth or filters between the space and ground.
- The noise threshold is a user input. This should be measured directly from the image.
- pixel scales should be read from fits headers rather than user input

```python
import numpy
import matplotlib.pyplot as plt
from scipy import ndimage
from scipy.ndimage import measurements, filters
import pyfits
import skimage
from skimage import filter
```

```python
# increase the notebook image size
figsize(10,8)
```

# User input

```python
# HST fits image
image_file_space = '../TestData/cl0916_f814w_drz_sci.fits'

# Ground based seeing
psf_ground_arcsec = 0.7 #arcsec

# Pixel scales of the detectors
pixscale_ground = 0.2 #arcsec/pix
pixscale_space = 0.05 #arcsec/pix
```

#Program

```python
# determine the size of the ground PSF convolution kernel in space pixel coordinates
psf_ground_spix = psf_ground_arcsec / pixscale_space # space pixels
print 'psf_ground corresponds to {0:0.1f} space pixels'.format(psf_ground_spix)
```

```python
# load the HST fits image
fits_hst = pyfits.open(image_file_space)
# print some information about the fits image
fits_hst.info()
```

There is just a single header and image frame in this image.

```python
# assign the image data to an array
data_hst = fits_hst[0].data
```

```python
img_hst = plt.imshow(data_hst,vmin=0,vmax=500)
```

```python
# convolve the space image with the ground seeing
# currently assumes just a Gaussian convolution
data_hst_conv = ndimage.filters.gaussian_filter(data_hst,psf_ground_spix,mode='constant')
```

```python
img_hst_conv = plt.imshow(data_hst_conv,vmin=0,vmax=500)
```

```python
# perform percolation 
img_hst_conv_seg, num_hst_conv_seg = ndimage.measurements.label(data_hst_conv)
```

```python
plt.imshow(img_hst_conv_seg,origin='lower',interpolation='nearest')
```

This was to be expected since measurements.label will link anything with non-zero pixels. Thus we need to process the image to determine a segmentation threshold first.

```python
thresh = filter.threshold_otsu(data_hst_conv)
```

```python
mask = data_hst_conv < thresh
```

```python
plt.imshow(mask)
```

## Estimating the noise

J. Immerkær, “Fast Noise Variance Estimation”, Computer Vision and Image Understanding, Vol. 64, No. 2, pp. 300-302, Sep. 1996

this method essentially uses a high-pass filter to isolate the noise

```python
# image dimensions
H, W = data_hst_conv.shape
# convolution filter
filter_fast = [[1, -2, 1],
               [-2, 4, -2],
               [1, -2, 1]]
# calculate the noise level in the image
sigma_space = numpy.sum(numpy.sum(numpy.absolute(ndimage.filters.convolve(data_hst, filter_fast,mode='constant'))))
sigma_space = sigma_space * numpy.sqrt(0.5 * numpy.pi) / (6 * (W-2) * (H-2))
```
